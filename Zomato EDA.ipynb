```python
df2 = pd.read_excel('Country-Code.xlsx')
```

```python
df2.head()
```

```python
df2.columns
```

```python
#combining both data frames to get the respective countries
final_df = pd.merge(df,df2,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
countries = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Top 3 countries 
plt.pie(country_values[:3],labels=countries[:3],autopct='%1.2f%%')
```

Conclusion

- Majority of the operations of Zomato is based in india, followed by US and UK

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Conclusion

- When the ratings are between 1.8 to 2.4 are Poor
- When the ratings are between 2.5 to 3.4 are Average
- When the ratings are between 3.5 to 3.9 are Good
- When the ratings are between 4.0 to 4.4 are Very Good
- When the ratings are between 4.5 to 4.9 are Excellent
- There are 2148 records without any rating (0.0) 

```python
plt.subplots(figsize=(15, 7))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
plt.subplots(figsize=(15, 7))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette = ['black','red','orange','yellow','green','green'])
```

Conclusions

- Maximum ratings are between 2.5 to 2.4
- 0 / Not rated count is very high

```python
sns.countplot(x='Rating color',data = ratings,palette = ['black','red','orange','yellow','green','green'] )
```

```python
# Q1. Countries that have given 0 ratings

final_df[df['Aggregate rating'] == 0].groupby('Country').size().reset_index().rename(columns={0:'Count'})
```

Conclusion

- Maximum Zero ratings is from Indian region

```python
# Q2. Currencies used by Countries

final_df[['Currency','Country']].groupby(['Country','Currency']).size().reset_index()
```

```python
# Q3. Countries with online delivery option
final_df[df['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index()
```

```python
# Top 3 Cities Ditribution

cities = final_df.City.value_counts().index
city_values = final_df.City.value_counts().values
```

```python
plt.pie(city_values[:3],labels = cities[:3],autopct='%1.2f%%')
```

Conclusion

- Maximum operations of Zomato in India are based in New Delhi
